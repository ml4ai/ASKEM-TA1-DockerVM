```python
from askem_extractions.data_model import AttributeCollection
from pathlib import Path
import requests

# Endpint configuration. Set the host correctly depending on where are you running this code snippet
# ENDPOINT="https://api.askem.lum.ai" # Uncomment this line to use SKEMA's web service
ENDPOINT = "http://127.0.0.1:8000" # Uncomment this line to use the docker compose service

def test_link_amr():
    """Test that we are able to link and AMR to text extractions correctly"""
    amr_path = Path("..") / "data" / "metal" / "fixed_scenario1_amr_eq.json"
    extractions_path = Path("..") / "data" / "metal" / "tr_document_results.json"

    params = {
        "similarity_model": "sentence-transformers/all-MiniLM-L6-v2",
        "similarity_threshold": 0.5
    }

    files = {
        "amr_file": amr_path.open("rb"),
        "text_extractions_file": extractions_path.open("rb")
    }

    response = requests.post(f"{ENDPOINT}/metal/link_amr", params=params, files=files)

    assert response.status_code == 200, f"The response code for /link_amr was {response.status_code}"

    linked_amr = response.json()
    
    return linked_amr
    
print(test_link_amr())
```
