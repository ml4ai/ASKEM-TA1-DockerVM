# Text Reading Integration Notebook

## Plain Text Annotation Example

```python
import requests
from pprint import pprint

# Endpint configuration. Set the host correctly depending on where are you running this code snippet
ENDPOINT="http://integrated-tr:7778"

# Read an example document to annotate
test_text = open("/data/skema/text/1-s2.0-S2211379721005490-main.txt").read()

# Set the annotation parameters, which pipelines do I want to run?
parameters = {
    "annotate_skema": True,
    "annotate_mit": False
}

# Create the request body's payload. It is an array of documents to annotate
documents_to_annotate = {
  "texts": [
    test_text,
      
  ]
}

# Call the REST API
response = requests.post(f"{ENDPOINT}/integrated_text_extractions/",
                         params=parameters, json=documents_to_annotate)

# Collect the response and display
if response.status_code == 200:
    print(response.json())
else:
    print(response)
```
